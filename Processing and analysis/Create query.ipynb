```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
df_movie_reshaped = pd.read_csv(r'C:\Users\misha\Pet-project\Processing and analysis\df_movie_reshaped.csv',lineterminator = '\n',sep = ';')
```

```python
df_movie_reshaped = df_movie_reshaped.rename(columns = {'RecommensationsBook\r':'RecommensationsBook',})
```

```python
with open("data_base_query1.txt","w+", encoding="utf-8") as my_file:
     my_file.write(f'INSERT INTO links {tuple(df_movie_reshaped.columns)}\n VALUES')
     for _, row in df_movie_reshaped.iterrows():
            output_str = "('{0}', '{1}', '{2}', '{3}', '{4}','{5}', '{6}', '{7}')".format(
            row['title'], row['score'], row['date'], row['summary'], row['Atmosphere'], row['Tags'], row['RecommensationsGame'], row['RecommensationsBook'])
            my_file.write(f'{output_str}\n')
```
